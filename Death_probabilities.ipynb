```python
import pandas as pd
import numpy as np
from bokeh.charts import HeatMap, output_file, show
from bokeh.plotting import figure
from bokeh.io import output_notebook, show

output_notebook()
```

```python
casualties = pd.read_csv('Casualties0515.csv', error_bad_lines= False)

casualties = casualties[['Age_Band_of_Casualty', 'Casualty_Severity']]

x = casualties.groupby(['Age_Band_of_Casualty']).size()

y = casualties[casualties.Casualty_Severity == 1].groupby(['Age_Band_of_Casualty']).size()

p = figure(plot_width=400, plot_height=400)

p.circle(casualties.Age_Band_of_Casualty.unique(), y / x, size=12, line_color="navy", fill_color="orange", fill_alpha=51)

show(p)
```
