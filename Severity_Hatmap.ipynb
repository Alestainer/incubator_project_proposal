```python
import pandas as pd
import numpy as np
from bokeh.charts import HeatMap, output_file, show
from bokeh.plotting import figure
from bokeh.io import output_notebook, show

output_notebook()
```

```python
df = pd.read_csv('Accidents0515.csv')

small_df = df[['Accident_Severity', 'Light_Conditions', 'Road_Surface_Conditions']]

small_df = small_df.groupby(['Light_Conditions', 'Road_Surface_Conditions'], as_index=False).mean()

sdf = list()
for i in (small_df.Light_Conditions.unique()):
    for j in (small_df.Road_Surface_Conditions.unique()):
        k = small_df.loc[small_df['Light_Conditions'] == i].loc[small_df['Road_Surface_Conditions'] == j].Accident_Severity.mean()
        sdf.append([i,j,k])

last_df = pd.DataFrame(sdf, columns=['light', 'road', 'severity'])
last_df = last_df.replace(-1, 0)

hm1 = HeatMap(last_df, x='light', y='road', values='severity', stat=None,
              sort_dim={'x': False}, width=900, plot_height=900)

show(hm1)
```
